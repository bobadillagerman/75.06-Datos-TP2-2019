```python
import numpy as np
import pandas as pd
import csv
from time import time

from datetime import datetime
import category_encoders as ce
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

```python
train = pd.read_csv('train.csv')
```

```python
trainCaras = train[(train['precio']>3350000)]
trainIntermedias = train[(train['precio']<2540000)&(train['precio']>1400000)]
trainBaratas = train[(train['precio']<940000)]
del train
```

```python
trainAux = trainCaras.copy()
trainAux.dropna(subset=["titulo"],inplace=True)
titulosCaras = trainAux["titulo"]

trainAux = trainIntermedias.copy()
trainAux.dropna(subset=["titulo"],inplace=True)
titulosIntermedias = trainAux["titulo"]

trainAux = trainBaratas.copy()
trainAux.dropna(subset=["titulo"],inplace=True)
titulosBaratas = trainAux["titulo"]

del trainAux
```

```python
trainAux = trainCaras.copy()
trainAux.dropna(subset=["descripcion"],inplace=True)
descripcionesCaras = trainAux["descripcion"]

trainAux = trainIntermedias.copy()
trainAux.dropna(subset=["descripcion"],inplace=True)
descripcionesIntermedias = trainAux["descripcion"]

trainAux = trainBaratas.copy()
trainAux.dropna(subset=["descripcion"],inplace=True)
descripcionesBaratas = trainAux["descripcion"]

del trainAux
del trainCaras
del trainIntermedias
del trainBaratas
```

```python
spanish_stopwords = stopwords.words('spanish')
stemmer = SnowballStemmer('spanish')

tipos = ['casa','apartamento','condominio','terreno','local','oficina','bodega','edificio','comercial','quinta','duplex','rancho','departamento','nave','industrial','lote','hospedaje','garage','venta','vende','mexico','dept','departamento','Casa','dept.','departamentos','Mexico','vendo','depto','depto.']
spanish_stopwords.extend(tipos)
```

```python
def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed
```

# Para Titulos

```python
palabrasBaratas=[]
for row in titulosBaratas:
    row =row.lower()
    row=filter(None, re.split("[, \-!?:._¡¿'`<>;()+]", row))
    for palabra in row:
        if len(palabra)<4: #saco conectores
            continue
        if palabra not in spanish_stopwords:
            palabrasBaratas.append(palabra)
```

```python
palabrasIntermedias=[]
for row in titulosIntermedias:
    row =row.lower()
    row=filter(None, re.split("[, \-!?:._¡¿'`<>;()+]", row))
    for palabra in row:
        if len(palabra)<4: #saco conectores
            continue
        if palabra not in spanish_stopwords:
            palabrasIntermedias.append(palabra)
```

```python
palabrasCaras=[]
for row in titulosCaras:
    row =row.lower()
    row=filter(None, re.split("[, \-!?:._¡¿'`<>;()+]", row))
    for palabra in row:
        if len(palabra)<4: #saco conectores
            continue
        if palabra not in spanish_stopwords:
            palabrasCaras.append(palabra)
```

```python
stems = stem_tokens(palabrasCaras, stemmer)
contador = Counter(stems)
palabrasComunesCaras = contador.most_common(100)
palabrasComunesCaras
```

```python
stems = stem_tokens(palabrasIntermedias, stemmer)
contador = Counter(stems)
palabrasComunesIntermedias = contador.most_common(100)
palabrasComunesIntermedias
```

```python
stems = stem_tokens(palabrasBaratas, stemmer)
contador = Counter(stems)
palabrasComunesBaratas = contador.most_common(100)
palabrasComunesBaratas
```
